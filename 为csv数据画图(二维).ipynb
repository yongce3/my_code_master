```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# 设文件路径
file_paths = [
    '../../Downloads/luna_vertical.csv',
    '../../Downloads/luna_horization.csv',
    '../../Downloads/LUNA_sim_H.csv',
    '../../Downloads/LUNA_sim_V.csv',
    '../../Downloads/JPL.csv',
    '../../Downloads/Default Dataset.csv'
    
]

 
# 颜色、标签和点标记列表  区分文件 
colors = ['b', 'g', 'r', 'c','purple','brown']
labels = ['LUNA Csedreki et al.,.2021(V) F=1.281',
          'LUNA Csedreki et al.,.2021(H) F=1.284',
          'LUNA Csedreki et al.,.2021(Hs) F=1.281',
          'LUNA Csedreki et al.,.2021(Vs)F=1.284',
          'JPL Li et al.,.2022 F=2.545',
          'CASPAR Falahat et al.,.2013 F=1.645']
markers = ['o', 's', 'd', '^','H','X']  

#画布
plt.figure(figsize=(8, 6))

for i, file_path in enumerate(file_paths):
    # 读CSV文件，中文逗号分隔符
    data = pd.read_csv(file_path, sep='，', engine='python')

    # 第一列是X，第二列是Y
    x = data.iloc[:, 0]
    y = data.iloc[:, 1]

    # 画散点图
    #plt.plot(x, y, color=colors[i], label=labels[i], marker=markers[i],linestyle='-')
    plt.scatter(x, y, color=colors[i], label=labels[i], marker=markers[i], alpha=1,  )
# 加legend
font = {'family': 'Arial', 'style': 'italic', 'weight': 'bold', 'size': 12}
plt.legend(fontsize=12,frameon=0,prop=font)



#美化
# 标题与坐标轴标签
plt.title('Detection Efficiency vs. Neutron Energy', fontsize=18,fontweight ='bold',pad=16)
plt.xlabel('Neutron Energy(MeV)', fontsize=16,fontweight ='bold',labelpad=10)
plt.ylabel('Detection Efficiency(%)', fontsize=16,fontweight ='bold',labelpad=10)
ax = plt.subplot (1,1,1)
ax.tick_params(direction='in', which='both', bottom=True, top=True, left=True, right=True, labelsize=10)
ax.tick_params(which='major', length=6, width=1.5)
ax.tick_params(which='minor', length=3, width=1)

# 设置x轴的主刻度线为0.5的倍数
ax.xaxis.set_major_locator(MultipleLocator(0.5))

# 设置x轴的次刻度线为0.1的倍数
ax.xaxis.set_minor_locator(MultipleLocator(0.1))

# 设置y轴的主刻度线为0.5的倍数
ax.yaxis.set_major_locator(MultipleLocator(5))

# 设置y轴的次刻度线为0.1的倍数
ax.yaxis.set_minor_locator(MultipleLocator(0.5))

ax.set_xlim(-0.3,5.3)
ax.set_ylim(15,55)

# plt.show()
plt.savefig('./效率曲线.pdf',dpi=600)
```
